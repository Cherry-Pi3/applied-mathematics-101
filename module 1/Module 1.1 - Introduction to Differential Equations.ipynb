## Why study them? <a id='ch2'></a>

This code accompanies the YouTube video [Differential Equations | Why study them? | Module 1.1.1](https://www.youtube.com/watch?v=DIXdOUeU75I&list=PLmKDBMy080MYDFyGWpVeV7QBxv6BSIqbT&index=4).

```python
'''
This block of code visualises the motion of a swinging pendulum as a sinusoidal wave over a time span of 2 seconds.

1. Parameters defining the pendulum's motion are set:
   - `A` is the amplitude, representing the maximum displacement from the equilibrium position.
   - `omega` is the angular frequency, calculated as 2 (since the period is assumed to be 1 second).
   - `phi` is the phase, set to 0, meaning the pendulum starts from its maximum displacement.

2. Time values are generated using `linspace`:
   - `t` represents time values ranging from 0 to 2 seconds, covering two complete cycles of the pendulum's motion.

3. The displacement values for each time `t` are computed using the sine function:
   - `y` represents the pendulum's displacement from its equilibrium position at each time `t`.

4. A figure is initialised with a specified size and a black background.

5. The pendulum's motion is plotted as a red curve, representing its displacement over time.

6. The plot is titled "Swinging Pendulum Motion" and labelled with white-coloured, appropriately sized axis labels.

7. The plot's appearance is further customised:
   - The background of the plotting area is set to black.
   - The axes are coloured white.
   - The top and right borders (spines) of the plot are hidden.
   - The tick labels on both axes are set to white and resized for clarity.

8. Finally, the plot is displayed using `plt.show()`.
'''

# 1. Define the parameters
A = 1  # Amplitude
omega = 2 * np.pi  # Angular frequency (period is assumed to be 1 second)
phi = 0  # Phase (pendulum starts from its maximum displacement)

# 2. Generate time values
t = np.linspace(0, 2, 400)  # 2 seconds (two cycles for a period of 1 second)

# 3. Compute the displacement values for each time t
y = A * np.sin(omega * t + phi)

# 4. Plotting with specified style
plt.figure(figsize=(8, 6), facecolor='white')  

# 5. The pendulum's motion is plotted as a red curve, representing its displacement over time.
plt.plot(t, y, color="red", linewidth=3)

# 6. Title and labels with specified font size and colour
plt.title("Swinging Pendulum Motion", color="black", fontsize=15)  
plt.xlabel("Time (seconds)", color="black", fontsize=12)  
plt.ylabel("Displacement from Equilibrium", color="black", fontsize=12)  

# 7. Setting white background and black axes
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black') 
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=12) 
plt.gca().tick_params(axis='y', colors='black', labelsize=12) 

# 8. Display plot
plt.show()
```

```python
'''
This code visualises the logistic growth of a population using a direction field and solution curves.

1. The `logistic_growth` function defines the logistic growth differential equation.

2. A grid of points is created for both time `t` and population `y` values:
   - `T` and `Y` are mesh grids representing combinations of time and population values.

3. The slopes (derivatives) at each grid point are computed using the logistic growth equation.

4. The direction field is represented using fixed-length dashes:
   - `dx` is a fixed change in the x-direction (time).
   - `dy` represents the change in the y-direction (population) at each grid point.
   - The lengths of the dashes are normalised to a fixed length, specified by `dash_length`.

5. The direction field is plotted as white dashes on a black background.

6. Solution curves for the logistic growth equation are plotted for different initial conditions:
   - `t_fine` provides a finer grid of time values for plotting smooth solution curves.
   - For each initial condition in `initial_conditions`, the logistic growth equation is solved using `odeint`, 
   and the solution curve is plotted in light green.

7. The plot is titled "Direction Field for Logistic Growth" and labelled with white-coloured, appropriately sized axis 
   labels.

8. The plot's appearance is further customised:
   - The background of the plotting area is set to black.
   - The axes are coloured white.
   - The top and right borders (spines) of the plot are hidden.
   - The tick labels on both axes are set to white and resized for clarity.

9. Finally, the plot is displayed using `plt.show()`.
'''

# 1. The `logistic_growth` function defines the logistic growth differential equation.
def logistic_growth(y, t, r, K):
    """Logistic growth differential equation."""
    return r * y * (1 - y/K)

# 2. A grid of points is created for both time `t` and population `y` values:
t = np.linspace(0, 100, 20)
y = np.linspace(0, 1500, 20)
T, Y = np.meshgrid(t, y)

# 3. The slopes (derivatives) at each grid point are computed using the logistic growth equation.
dy = logistic_growth(Y, T, 0.1, 1000)
dx = np.ones(dy.shape)  

# 4. The direction field is represented using fixed-length dashes:
length = np.sqrt(dx**2 + dy**2)
dx /= length
dy /= length

# Set the fixed length for the dashes
dash_length = 15
dx *= dash_length
dy *= dash_length

# 5. The direction field is plotted as black dashes on a white background.
plt.figure(figsize=(8, 6), facecolor='white')  
for i in range(len(t)):
    for j in range(len(y)):
        plt.plot([T[i,j] - dx[i,j]/2, T[i,j] + dx[i,j]/2], 
                 [Y[i,j] - dy[i,j]/2, Y[i,j] + dy[i,j]/2], color='black')  

# 6. Solution curves for different initial conditions
t_fine = np.linspace(0, 100, 400)
initial_conditions = [100, 500, 1500]
for y0 in initial_conditions:
    y_sol = odeint(logistic_growth, y0, t_fine, args=(0.1, 1000))
    plt.plot(t_fine, y_sol, color='red', linewidth=2) 

# 7. Title and labels with specified font size and colour
plt.title("Direction Field for Logistic Growth", color="black", fontsize=15)  
plt.xlabel("Time", color="black", fontsize=12)  
plt.ylabel("Population", color="black", fontsize=12)  

# 8. The plot's appearance is further customised:
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black') 
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=12)  
plt.gca().tick_params(axis='y', colors='black', labelsize=12)  

# 9. Finally, the plot is displayed using `plt.show()`.
plt.show()
```

## Ordinary DEs <a id='ch3'></a>

This code accompanies the YouTube video [Differential Equations | Ordinary vs. Partial | Module 1.1.2](https://youtu.be/y-9SYkBkrjQ).

```python
'''
This code visualises the solution to the first-order ordinary differential equation (ODE) dy/dx = x + y.

1. The `ode_system` function defines the differential equation dy/dx = x + y.

2. An initial condition `y0` is specified, and a range of `x` values is defined using `linspace`.

3. The ODE is solved using the `odeint` function, which returns the solution `y` for the given `x` values.

4. The solution is plotted on a black background:
   - The curve representing the solution is plotted in red with a specified linewidth.
   - The title of the plot displays the differential equation being visualised.
   - The x and y axes are labelled with white-coloured, appropriately sized labels.

5. The plot's appearance is further customised:
   - The background of the plotting area is set to black.
   - The axes are coloured white.
   - The top and right borders (spines) of the plot are hidden.
   - The tick labels on both axes are set to white and resized for clarity.

6. Finally, the plot is displayed using `plt.show()`.
'''

# 1. The `ode_system` function defines the differential equation dy/dx = x + y.
def ode_system(y, x):
    """System of ODEs: dy/dx = x + y."""
    return x + y

# 2. An initial condition `y0` is specified, and a range of `x` values is defined using `linspace`.
y0 = [0]
x = np.linspace(0, 5, 100)

# 3. The ODE is solved using the `odeint` function, which returns the solution `y` for the given `x` values.
y = odeint(ode_system, y0, x)

# 4. The solution is plotted on a white background:
plt.figure(figsize=(8, 6), facecolor='white')  
plt.plot(x, y, color='red', linewidth=3)

# Title and labels with specified font size and color
plt.title(r"Solution to $\frac{dy}{dx} = x + y$", color="black", fontsize=15)  
plt.xlabel(r"$x$", color="black", fontsize=12)  
plt.ylabel(r"$y$", color="black", fontsize=12)  

# 5. The plot's appearance is further customised:
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black')  
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=12)  
plt.gca().tick_params(axis='y', colors='black', labelsize=12)  

# 6. Finally, the plot is displayed using `plt.show()`.
plt.show()
```

## Partial DEs <a id='ch4'></a>

```python
'''
This code simulates and visualises the heat distribution on a metal plate over time.

1. Parameters:
   - `alpha` represents the thermal diffusivity of the metal.
   - `x` and `y` are arrays of linearly spaced values representing the spatial dimensions of the plate.
   - `X` and `Y` are mesh grids created from `x` and `y` to represent the 2D spatial domain.

2. Initial Condition:
   - A Gaussian temperature distribution is set as the initial condition on the plate.
   - `sigma` is the standard deviation of the Gaussian distribution, which is increased to spread the heat more widely.

3. Heat Diffusion Simulation:
   - The `heat_diffusion` function simulates the heat diffusion over time using a finite difference method.
   - The function computes the temperature at each grid point based on its neighbors and the thermal diffusivity.
   - The simulation is run for a specified number of time steps (`num_steps`) with a given time step size (`dt`).

4. Plotting:
   - A contour plot is created to visualise the heat distribution on the plate.
   - The temperature distribution is represented using a colourmap (`cmap='hot'`), with 50 contour levels.
   - The plot is styled with a black background, white axes, and specified font sizes for the title and labels.
   - A colour bar (`colorbar`) is added to the plot to indicate the temperature scale.

5. The final heat distribution on the metal plate is displayed using `plt.show()`.
'''

# 1. Parameters
alpha = 0.01  # thermal diffusivity
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# 2. Initial condition: Gaussian temperature distribution with larger standard deviation
sigma = 2.5  # increased standard deviation for wider spread
u_initial = np.exp(-X**2 / (2*sigma**2) - Y**2 / (2*sigma**2))

# 3. Simulate heat diffusion over time using a simple finite difference method
def heat_diffusion(u, dt, dx, dy, alpha):
    u_new = u.copy()
    for i in range(1, u.shape[0]-1):
        for j in range(1, u.shape[1]-1):
            u_new[i, j] = u[i, j] + alpha * dt * (
                (u[i+1, j] - 2*u[i, j] + u[i-1, j]) / dx**2 +
                (u[i, j+1] - 2*u[i, j] + u[i, j-1]) / dy**2
            )
    return u_new

# Simulate for a certain number of time steps
num_steps = 100
dt = 0.1
for _ in range(num_steps):
    u_initial = heat_diffusion(u_initial, dt, x[1]-x[0], y[1]-y[0], alpha)

# 4. Plotting with specified style
plt.figure(figsize=(8, 6), facecolor='white')  
contour = plt.contourf(X, Y, u_initial, cmap='hot', levels=50)

# Title and labels with specified font size and color
plt.title("Heat Distribution on a Metal Plate", color="black", fontsize=15)  
plt.xlabel(r"$x$", color="black", fontsize=12)  
plt.ylabel(r"$y$", color="black", fontsize=12)  

# Setting white background and black axes
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black')  
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=12)  
plt.gca().tick_params(axis='y', colors='black', labelsize=12)  

# Add a colour bar to the plot
plt.colorbar(contour)

# 5. Display plot
plt.show()
```

## Order of differential equations <a id='ch5'></a>

This code accompanies the YouTube video [Differential Equations | What is the Order? | Module 1.1.3](https://youtu.be/JspC0J3Xu0Q).

```python
'''
This code simulates and visualises the solution to a second-order ordinary differential equation (ODE) given by:
    d^2y/dx^2 + y = 0

1. System of ODEs:
   - The second-order ODE is transformed into a system of two first-order ODEs.
   - `Y[0]` represents y, and `Y[1]` represents v (dy/dx).
   - The `system` function computes the derivatives dy/dx and d^2y/dx^2 based on the current values of y and v.

2. Initial Conditions:
   - The initial conditions are set as y(0) = 1 and dy/dx(0) = 0.
   - `Y0` is a list containing these initial conditions.

3. Solution:
   - The `odeint` function from SciPy is used to solve the system of ODEs over a specified range of x values.
   - The solution for y is extracted from the resulting array.

4. Plotting:
   - The solution y is plotted against x to visualise the behaviour of the system over time.
   - The plot is styled with a black background, white axes, and specified font sizes for the title and labels.

5. The final solution curve is displayed using `plt.show()`.
'''

# 1. System of ODEs: Y[0] is y and Y[1] is v (dy/dx)
def system(Y, x):
    y, v = Y
    dydx = v
    dvdx = -y
    return [dydx, dvdx]

# 2. Initial conditions: y(0) = 1, dy/dx(0) = 0
Y0 = [1, 0]
x = np.linspace(0, 10, 200)

# 3. Solve the system of ODEs
sol = odeint(system, Y0, x)
y = sol[:, 0]

# 4. Plotting with specified style
plt.figure(figsize=(8, 6), facecolor='white')  
plt.plot(x, y, color='red', linewidth=3)

# Title and labels with specified font size and color
plt.title(r"Solution to $\frac{d^2y}{dx^2} + y = 0$", color="black", fontsize=15) 
plt.xlabel(r"$x$", color="black", fontsize=12)  
plt.ylabel(r"$y$", color="black", fontsize=12)  

# Setting white background and black axes
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black')  
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=12)  
plt.gca().tick_params(axis='y', colors='black', labelsize=12)  

# 5. Display plot
plt.show()
```

## Solutions and initial conditions <a id='ch7'></a>

This code accompanies the YouTube video [Differential Equations | Solutions and initial conditions | Module 1.1.5](https://youtu.be/SH_S2feVeqM).

```python
'''
This code visualises the solutions to a first-order linear ordinary differential equation (ODE) given by dy/dx = -ry, 
where 'r' is a constant.

1. Differential Equation:
   - The ODE is represented by the function `solution`, which returns the analytical solution y(x) = C * exp(-r * x).
   - 'C' is an integration constant that represents the initial condition y(0) = C.
   - The default value for the constant 'r' is set to 1.

2. x values:
   - A range of x values is defined using `np.linspace` from 0 to 5.

3. Different Initial Conditions:
   - A list of initial conditions (values of C) is defined.
   - For each initial condition, the solution y(x) is computed and plotted against x.

4. Plotting:
   - The solutions for each initial condition are plotted on the same graph to compare their behaviours.
   - Each solution curve is labelled with its corresponding initial condition value 'C'.
   - The plot is styled with a black background, white axes, and specified font sizes for the title, labels, and legend.

5. The final comparison plot, showcasing the solutions for different initial conditions, is displayed using 
   `plt.show()`.
'''

# 1. Differential equation: dy/dx = -ry
def solution(x, C, r=1):
    return C * np.exp(-r * x)

# 2. x values
x = np.linspace(0, 5, 200)

# 3. Different initial conditions (values of C)
initial_conditions = [1, 2, 3, 4, 5]

# 4. Plotting with specified style
plt.figure(figsize=(8, 6), facecolor='white')  
for C in initial_conditions:
    plt.plot(x, solution(x, C), linewidth=3, label=f'C = {C}')

# Title, labels, and legend with specified font size and color
plt.title("Solutions for Different Initial Conditions", color="black", fontsize=15)  
plt.xlabel(r"$x$", color="black", fontsize=12)  
plt.ylabel("$y$", color="black", fontsize=12)  
plt.legend(loc='upper right', fontsize=12, facecolor='white', edgecolor='black', labelcolor='black')  

# Setting white background and black axes
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black')  
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=12)  
plt.gca().tick_params(axis='y', colors='black', labelsize=12)  

# 5. Display plot
plt.show()
```

## Methods of solution <a id='ch8'></a>

This code accompanies the YouTube video [Differential Equations | Methods of Solution | Module 1.1.6](https://youtu.be/bp7H3IAiuzg).

```python
'''
1. Analytical Solution:
   - The function `analytical_solution` provides the exact solution y(x) = exp(x) to the differential equation.

2. Euler's Method:
   - The function `euler_method` approximates the solution using Euler's method, a simple numerical method for solving 
     ordinary differential equations (ODEs).
   - Given an initial value y0, the function computes the solution at discrete points in the interval defined by 'x'.

3. Differential Equation:
   - The ODE dy/dx = y is represented by the function `f`.

4. x values:
   - Two sets of x values are defined: a coarse grid for Euler's method and a fine grid for the analytical solution.

5. Solution Computation:
   - The solutions are computed for both the analytical method and Euler's method using the defined x values.

6. Plotting:
   - The solutions are plotted on the same graph to compare the exact analytical solution with the approximation from 
     Euler's method.
   - The analytical solution is plotted as a solid red line, while Euler's method is plotted as a dashed light blue 
     line.
   - The plot is styled with a black background, white axes, and specified font sizes for the title, labels, and legend.

7. The final comparison plot, showcasing the analytical solution versus Euler's method, is displayed using `plt.show()`.
'''

# 1. Analytical solution
def analytical_solution(x):
    return np.exp(x)

# 2. Euler's method
def euler_method(f, y0, x):
    h = x[1] - x[0]
    y = [y0]
    for i in range(len(x) - 1):
        y_next = y[-1] + h * f(y[-1], x[i])
        y.append(y_next)
    return np.array(y)

# 3. Differential equation
def f(y, x):
    return y

# 4. x values for Euler's method (coarse grid) and for analytical solution (fine grid)
x_euler = np.linspace(0, 5, 10)
x_analytical = np.linspace(0, 5, 500)

# 5. Analytical solution and Euler's method approximation
y_analytical = analytical_solution(x_analytical)
y_euler = euler_method(f, 1, x_euler)

# 6. Plotting with specified style
plt.figure(figsize=(8, 6), facecolor='white')  # Changed background to white
plt.plot(x_analytical, y_analytical, color='red', linewidth=3, label='Analytical Solution')
plt.plot(x_euler, y_euler, color='blue', linestyle='--', linewidth=2, label="Euler's Method")  

# Title, labels, and legend with specified font size and color
plt.title("Analytical vs. Numerical Solution", color="black", fontsize=15)  
plt.xlabel(r"$x$", color="black", fontsize=12)  
plt.ylabel(r"$y$", color="black", fontsize=12)  
plt.legend(loc='upper left', fontsize=12, facecolor='white', edgecolor='black', labelcolor='black')  

# Setting white background and black axes
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black')  
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=12)  
plt.gca().tick_params(axis='y', colors='black', labelsize=12)  

# 7. Set limits for x- and y-axis and generate plot
plt.xlim([0, 5])
plt.ylim([0, np.exp(5)])
plt.show()
```
