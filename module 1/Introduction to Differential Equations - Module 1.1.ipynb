## What are they? <a id='ch1'></a>

This code accompanies the YouTube video [Differential Equations | What are they? | Module 1.1.0](https://www.youtube.com/watch?v=0fqG2XA7Sl8&list=PLmKDBMy080MYDFyGWpVeV7QBxv6BSIqbT&index=3).

```python
'''
This block of code visualises the exponential function over the interval [-2, 2] with a special emphasis on the point 
x=0.

1. Two sets of x-values are generated:
   - `x1` covers the interval [-2, 0] and is used for the left half of the plot.
   - `x2` covers the interval [0, 2] and is used for the right half of the plot.

2. The exponential values for each set of x-values are computed:
   - `y1` represents the exponential values for `x1`.
   - `y2` represents the exponential values for `x2`.

3. A figure is initialised with a specified size and a black background.

4. The exponential function is plotted in two segments:
   - The left segment (for `x1`) is plotted as a dashed red line.
   - The right segment (for `x2`) is plotted as a solid red line.

5. A vertical turquoise dashed line is plotted at x=0, extending up to the point (0, e^0).

6. A horizontal turquoise dashed line is plotted at y=e^0, extending from x=-2 to x=0.

7. The plot is titled "Exponential Function: e^x" and labelled with white-coloured, appropriately sized axis labels.

8. The plot's appearance is further customised:
   - The background of the plotting area is set to black.
   - The axes are coloured white.
   - The top and right borders (spines) of the plot are hidden.
   - The tick labels on both axes are set to white and resized for clarity.

9. Finally, the plot is displayed using `plt.show()`.
'''

# 1. Generate x values for the two intervals
x1 = np.linspace(-2, 0, 200)
x2 = np.linspace(0, 2, 200)

# 2. Compute the exponential values for each x interval
y1 = np.exp(x1)
y2 = np.exp(x2)

# 3. Plotting
plt.figure(figsize=(15, 10), facecolor='white') 

# 4. Plotting the two line segments
plt.plot(x1, y1, color="red", linewidth=5, linestyle='--')  # Dashed line for [-2, 0]
plt.plot(x2, y2, color="red", linewidth=5)  # Solid line for [0, 2]

# 5. Plotting the vertical line (dashed and blue)
plt.vlines(0, 0, np.exp(0), color="black", linewidth=2, linestyles='--') 

# 6. Plotting the horizontal line
plt.hlines(np.exp(0), -2, 0, color="black", linewidth=2, linestyles='--') 

# 7. Title and labels with increased font size
plt.title(r"Exponential Function: $e^x$", color="black", fontsize=30)  
plt.xlabel(r"$x$", color="black", fontsize=20) 
plt.ylabel(r"$y$", color="black", fontsize=20) 

# 8. Setting white background and black axes
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black')  
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=20)  
plt.gca().tick_params(axis='y', colors='black', labelsize=20)  

# 9. Display plot
plt.show()
```

## Why study them? <a id='ch2'></a>

This code accompanies the YouTube video [Differential Equations | Why study them? | Module 1.1.1](https://www.youtube.com/watch?v=DIXdOUeU75I&list=PLmKDBMy080MYDFyGWpVeV7QBxv6BSIqbT&index=4).

```python
'''
This block of code visualises the motion of a swinging pendulum as a sinusoidal wave over a time span of 2 seconds.

1. Parameters defining the pendulum's motion are set:
   - `A` is the amplitude, representing the maximum displacement from the equilibrium position.
   - `omega` is the angular frequency, calculated as 2π (since the period is assumed to be 1 second).
   - `phi` is the phase, set to 0, meaning the pendulum starts from its maximum displacement.

2. Time values are generated using `linspace`:
   - `t` represents time values ranging from 0 to 2 seconds, covering two complete cycles of the pendulum's motion.

3. The displacement values for each time `t` are computed using the sine function:
   - `y` represents the pendulum's displacement from its equilibrium position at each time `t`.

4. A figure is initialised with a specified size and a black background.

5. The pendulum's motion is plotted as a red curve, representing its displacement over time.

6. The plot is titled "Swinging Pendulum Motion" and labelled with white-coloured, appropriately sized axis labels.

7. The plot's appearance is further customised:
   - The background of the plotting area is set to black.
   - The axes are coloured white.
   - The top and right borders (spines) of the plot are hidden.
   - The tick labels on both axes are set to white and resized for clarity.

8. Finally, the plot is displayed using `plt.show()`.
'''

# 1. Define the parameters
A = 1  # Amplitude
omega = 2 * np.pi  # Angular frequency (period is assumed to be 1 second)
phi = 0  # Phase (pendulum starts from its maximum displacement)

# 2. Generate time values
t = np.linspace(0, 2, 400)  # 2 seconds (two cycles for a period of 1 second)

# 3. Compute the displacement values for each time t
y = A * np.sin(omega * t + phi)

# 4. Plotting with specified style
plt.figure(figsize=(15, 10), facecolor='white')  

# 5. The pendulum's motion is plotted as a red curve, representing its displacement over time.
plt.plot(t, y, color="red", linewidth=5)

# 6. Title and labels with specified font size and colour
plt.title("Swinging Pendulum Motion", color="black", fontsize=30)  
plt.xlabel("Time (seconds)", color="black", fontsize=20)  
plt.ylabel("Displacement from Equilibrium", color="black", fontsize=20)  

# 7. Setting white background and black axes
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black') 
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=20) 
plt.gca().tick_params(axis='y', colors='black', labelsize=20) 

# 8. Display plot
plt.show()
```

```python
'''
This code visualises the logistic growth of a population using a direction field and solution curves.

1. The `logistic_growth` function defines the logistic growth differential equation.

2. A grid of points is created for both time `t` and population `y` values:
   - `T` and `Y` are mesh grids representing combinations of time and population values.

3. The slopes (derivatives) at each grid point are computed using the logistic growth equation.

4. The direction field is represented using fixed-length dashes:
   - `dx` is a fixed change in the x-direction (time).
   - `dy` represents the change in the y-direction (population) at each grid point.
   - The lengths of the dashes are normalised to a fixed length, specified by `dash_length`.

5. The direction field is plotted as white dashes on a black background.

6. Solution curves for the logistic growth equation are plotted for different initial conditions:
   - `t_fine` provides a finer grid of time values for plotting smooth solution curves.
   - For each initial condition in `initial_conditions`, the logistic growth equation is solved using `odeint`, 
   and the solution curve is plotted in light green.

7. The plot is titled "Direction Field for Logistic Growth" and labelled with white-coloured, appropriately sized axis 
   labels.

8. The plot's appearance is further customised:
   - The background of the plotting area is set to black.
   - The axes are coloured white.
   - The top and right borders (spines) of the plot are hidden.
   - The tick labels on both axes are set to white and resized for clarity.

9. Finally, the plot is displayed using `plt.show()`.
'''

# 1. The `logistic_growth` function defines the logistic growth differential equation.
def logistic_growth(y, t, r, K):
    """Logistic growth differential equation."""
    return r * y * (1 - y/K)

# 2. A grid of points is created for both time `t` and population `y` values:
t = np.linspace(0, 100, 20)
y = np.linspace(0, 1500, 20)
T, Y = np.meshgrid(t, y)

# 3. The slopes (derivatives) at each grid point are computed using the logistic growth equation.
dy = logistic_growth(Y, T, 0.1, 1000)
dx = np.ones(dy.shape)  

# 4. The direction field is represented using fixed-length dashes:
length = np.sqrt(dx**2 + dy**2)
dx /= length
dy /= length

# Set the fixed length for the dashes
dash_length = 15
dx *= dash_length
dy *= dash_length

# 5. The direction field is plotted as black dashes on a white background.
plt.figure(figsize=(20, 12), facecolor='white')  
for i in range(len(t)):
    for j in range(len(y)):
        plt.plot([T[i,j] - dx[i,j]/2, T[i,j] + dx[i,j]/2], 
                 [Y[i,j] - dy[i,j]/2, Y[i,j] + dy[i,j]/2], color='black')  

# 6. Solution curves for different initial conditions
t_fine = np.linspace(0, 100, 400)
initial_conditions = [100, 500, 1500]
for y0 in initial_conditions:
    y_sol = odeint(logistic_growth, y0, t_fine, args=(0.1, 1000))
    plt.plot(t_fine, y_sol, color='red', linewidth=3) 

# 7. Title and labels with specified font size and colour
plt.title("Direction Field for Logistic Growth", color="black", fontsize=30)  
plt.xlabel("Time", color="black", fontsize=20)  
plt.ylabel("Population", color="black", fontsize=20)  

# 8. The plot's appearance is further customised:
plt.gca().set_facecolor('white')  
plt.gca().spines['left'].set_color('black') 
plt.gca().spines['bottom'].set_color('black')  
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().tick_params(axis='x', colors='black', labelsize=20)  
plt.gca().tick_params(axis='y', colors='black', labelsize=20)  

# 9. Finally, the plot is displayed using `plt.show()`.
plt.show()
```
